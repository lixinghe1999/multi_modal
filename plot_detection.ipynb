```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# VoxelNet-based
# latency: read_data, load_gpu, inference
# latency_model: VFE, BACKBONE3D, TO_BEV, BACKBONE2D, DENSE_HEAD, POST_PROCESSING
fig, axs = plt.subplots(2, 1)
# normal
latency = [1.26, 35.43, 35.67]
latency_model = [0.21, 16.27, 1.2, 15.6, 2.37, 3.25]
axs[0].bar(np.arange(len(latency))+0.125, latency, width=0.25, color='r')
axs[1].bar(np.arange(len(latency_model))+0.125, latency_model, width=0.25, color='r')
# 4 * downsample, we observe drop in gpu loading but increase for inference
latency = [2.8, 5.9, 47.18]
latency_model = [0.25, 19, 1.27, 16, 2.6, 3.1]
axs[0].bar(np.arange(len(latency))-0.125, latency, width=0.25, color='b')
axs[1].bar(np.arange(len(latency_model))-0.125, latency_model, width=0.25, color='b')
axs[0].set_xticks(np.arange(len(latency)), ['read_data', 'load_gpu', 'inference'])
axs[1].set_xticks(np.arange(len(latency_model)), ['VFE', 'BACKBONE3D', 'BEV', 'BACKBONE2D', 'HEAD', 'POST'])
plt.show()
```

```python
# PointNet-based
# latency: read_data, load_gpu, inference
# latency_model: BACKBONE3D, POINT_HEAD, ROI, POST_PROCESSING
fig, axs = plt.subplots(2, 1)
# normal
latency = [0.2, 1.6, 85]
latency_model = [48, 1.8, 33.8, 2.1]
axs[0].bar(np.arange(len(latency))+0.125, latency, width=0.25, color='r')
axs[1].bar(np.arange(len(latency_model))+0.125, latency_model, width=0.25, color='r')
# 4 * downsample, we observe drop in gpu loading but increase for inference
latency = [0.2, 2, 56]
latency_model = [31.7, 1.5, 21, 2.6]
axs[0].bar(np.arange(len(latency))-0.125, latency, width=0.25, color='b')
axs[1].bar(np.arange(len(latency_model))-0.125, latency_model, width=0.25, color='b')
axs[0].set_xticks(np.arange(len(latency)), ['read_data', 'load_gpu', 'inference'])
axs[1].set_xticks(np.arange(len(latency_model)), ['BACKBONE3D', 'POINT_HEAD', 'ROI', 'POST'])
plt.show()
```
